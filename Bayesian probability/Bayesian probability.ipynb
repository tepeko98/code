```python
# 자연어 처리에서 유용한 코드!

def process_text(text):

  # text에서 구두점(punctuation)을 제거
  nopunc = [char for char in text if char not in string.punctuation] # text에서 하나씩 뽑아서 char에 넣고 char에 구두점이 없다면 nopunc에 넣어라
  nopunc = ''.join(nopunc)

  # text에서 불용어 제거
  cleaned_words = [word for word in nopunc.split()
                  if word.lower() not in stopwords.words('english')]

  return cleaned_words
```

```python
import nltk
nltk.download('stopwords')
```

```python
df['text'].apply(process_text)
```

```python
# 행렬 변환

from sklearn.feature_extraction.text import CountVectorizer # count를 세어서 vector를 뽑아주는 패키지
msg_bow = CountVectorizer(analyzer = process_text).fit_transform(df['text']) # 한 번에 단어주머니(back of word)를 만들었다.
```

```python
msg_bow
```

```python
# data split (train : 80%, test : 20%)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(msg_bow, df['label_num'], test_size=0.2, random_state=0)
```

```python
# 다항식 나이브메이즈 모델 훈련

from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
```
