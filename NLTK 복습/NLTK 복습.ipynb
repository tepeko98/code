```python
example = "hi i`m ga-yeon. nice to meet you. I`m so happy"
```

```python
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
ex = 'drwill is very import person. It`s everything' 
stop_words = set(stopwords.words('english'))
```

```python
word_tokens = word_tokenize(example)

result = [ ]

for w in word_tokens:
  if w not in stop_words:
    result.append(w)

print('원문', word_tokens)
print('불용어 제거문', result)
```

```python
example2 = "아무말이나 해보자. 무슨 말을 할까. 어디서부터 어떻게. 재밌네요."
stop_words2 = "아냐, 알지"

stop_words2 = stop_words2.split(' ')

word_tokens2 = word_tokenize(example2)

result = [w for w in word_tokens2 if not w in stop_words2]
```

```python
# 정수 인코딩

text = "I've been havin' dreams \
        Jumpin' on a trampoline \
        Flippin' in the air \
        I never land, just float there \
        As I'm looking up \
        Suddenly the sky erupts"
```

```python
# sentence tokenization
from nltk.tokenize import sent_tokenize
stop_words = set(stopwords.words('english'))

sent_tokenize(text)
```

```python
print(text)
```

```python
# word tokenization

sent = [ ]
stop_words = set(stopwords.words('english')) # nltk의 불용어

for i in text:
  sent = word_tokenize(i)
  result = [ ]

  for w in sent:
    w = w.lower()
    if w not in stop_words:
      if len(w) > 2:
        result.append(w)
  sent.append(result)

print(sent)
```
