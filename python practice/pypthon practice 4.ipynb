```python
def fact(n):
    output = 1
    for i in range(1, n+1):
        output *= i
    return output
print("2!: ", fact(2))
print("3!: ", fact(3))
print("4!: ", fact(4))
print("5!: ", fact(5))
```

```python
def facto(n):
    if n == 0:
        return 1
    else:
        return n * facto(n-1)
    
print("2!: ", facto(2))
print("3!: ", facto(3))
print("4!: ", facto(4))
print("5!: ", facto(5))
```

```python
def factor(n):
    if n == 0:
        return 1
    else:
        return n * factor(n - 1)
    
print("2!: ", factor(2))
print("3!: ", factor(3))
print("4!: ", factor(4))
print("5!: ", factor(5))
```

```python
def factorial(n):
    
    output = 1
    for i in range(1, n+1): #range(start,end+1,step)
        output *= i
    return output
print("4!: ", factorial(4))
```

```python
dict = {
    1: 1,
    2: 1
}
def fibo(n):
    if n in dict:
        return dict[n]
    else:
        output = fibo(n - 1) + fibo(n - 2)
        dict[n] = output
        return output
    
print(fibo(10))
        
```

```python
def power(item):
    return item * item

def under_19(item):
    return item < 19

list_a = [10, 15, 20, 30, 40]

output_a = map(power, list_a) #map(함수, 리스트)
print(output_a)
print(list(output_a))
print()

output_b = filter(under_19, list_a)
print(output_b)
print(list(output_b))
```

```python
#람다를 변수에 지정
power = lambda x: x * x
under_19 = lambda x: x < 19

list_b = [10,15,20,25,30]

output_c = map(power, list_b)
output_d = filter(under_19, list_b)
print(list(output_c))
print(list(output_d))
```

```python
import random

hangul = list("가나다라마바사아자차카타파하")

with open("names.txt", "w") as file:
    for i in range(1000):
        name = random.choice(hangul) + random.choice(hangul)
        weight = random.randrange(60, 90)#randrange로 몸무게 범위 조정
        height = random.randrange(160, 190)
        
        file.write("{},{},{}\n".format(name,weight,height))
        
with open("names.txt", "r") as file:
    names = file.read()
print(names)
```

```python
with open("names.txt", "r") as file:
    for line in file:
        (name, weight, height) = line.strip().split(",") #,로 구분 짓고, 자름
        
        if (not name) or (not weight) or (not height):
            continue #조건에 해당하는 게 나올 때까지 계속 반복하라
        
        bmi = int(weight) / (int(height) ** 2) * 10000
        
        
        result = ""
        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상인"
        else:
            result = "저체중"
            
        print('\n'.join([
            "이름: {}",
            "몸무게: {}",
            "키: {}",
            "bmi: {}",
            "결과: {}"
        ]).format(name, weight, height, bmi, result))
        print()
```

```python
def test():
    print("A지점 통과")
    yield 1
    print("B지점 통과")
    yield 2
    print("C지점 통과")

output = test()
print("D지점 통과")
print(next(output))
print("E지점 통과")
print(next(output))
print("F지점 통과")
print(next(output)) #마지막이기 때문에 StopIteration:이 뜬다.
```

```python
#변수 선언
list_all = ["52", "273", "32", "스파이", "103"]
#반복적용
list_num = []
for item in list_all:
    
    print(item) #나열하는거 걍 넣어봄
    
    #try 안에 넣고 숫자로 변환
    try:
        float(item)
        list_num.append(item)
    except:
        pass
print("{}안에 있는 숫자는".format(list_all))
print("{}입니다.".format(list_num))
```

```python
#함수 선언
def write_text_file(filename, text):
    #try except 구문 사용
    try:
        #파일 열기
        file = open(filename, "w")
        #여러 가지 처리 수행
        return
        #파일에 텍스트 입력
        file.write(text)
    except Exception as e:
        print(e)
    finally:
        #파일 닫기
        file.close()
#함수 호출        
write_text_file("test.txt", "안녕하세요!")
```

```python
list = [52,273,32,72,100]

try:
    num_input = int(input("정수 입력> "))
    
    print("{}번째 요소 {}: ".format(num_input, list[num_input]))
    
except ValueError as exception:
    print("정수를 입력해 주세요!")
    print("exception: ", exception)
    
except IndexError as exception:
    print("리스트의 인덱스를 벗어났어요!")
    print("exception: ", exception)
```
