matplotlib으로 데이터 시각화하기 1

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#통계를 표시하는 그래프 그리기
#groupby로 특정 부분을 묶은 뒤 그래프로 그린다.
iris.groupby(iris.species).mean()
```

```python
df = iris.groupby(iris.species).mean()
```

```python
#plot() 기본 선 그래프
df.plot()
```

```python
#plot.bar() 선그래프
df.plot.bar()
```

```python
#중간값으로 그래프 그리기
df2 = iris.groupby(iris.species).median()
df2.plot()
```

```python
df2.plot.bar()
```

```python
#종 이름을 가로로 바꾸려면 rot = 0 활용
df2.plot.bar(rot = 0)
```

```python
#산점도 그리기 scatter(x, y축 지정해줘야 함)
iris.plot.scatter(x='sepal_length', y='sepal_width')
```

```python
#그래프에 색칠하기 np.where(bool) 활용
x = np.where(iris['species'] == 'setosa', 'red', 'green')
iris.plot.scatter(x='sepal_length', y='sepal_width', c = x)
```

```python
#조건 확장
x = np.where(iris['species'] == 'setosa', 'red',
    np.where(iris['species'] == 'versicolor', 'green', 'blue'))
iris.plot.scatter(x='sepal_length', y='sepal_width', c = x)
```

```python
#x, y값만 바꿔서 다른 값 표시 가능
iris.plot.scatter(x='petal_length', y='petal_width', c = x)
```

seaborn으로 데이터 시각화하기 1

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#pairplot(data = iris) : 데이터의 행, 열을 순서대로 산점도 그려줌
#같은 경우 선 형태라 의미가 없으므로 히스토그램으로 나온다.
sns.pairplot(data = iris)
```

```python
#hue = '' : 기준점을 세우고 색깔구별을 해라
sns.pairplot(data = iris, hue = 'species')
```

```python
#특정값만 보고 싶을 때 vars = ['','']
sns.pairplot(data = iris, hue = 'species', vars = ['petal_length', 'petal_width'])
```

```python
#그래프 크기 지정 height = 4
sns.pairplot(data = iris, hue = 'species', vars = ['petal_length', 'petal_width'], height = 4)
```

```python
#jointplot : 두 개 섞어서 보여주는 그래프, kind는 종류 정하기
#scatter하면 안에 산점도, 밖에 히스토그램 나옴
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind = 'scatter')
```

```python
#reg : 경향을 선으로 나타내는 산점도 그래프
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind = 'reg')
```

```python
#kde: 확률 밀도
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind = 'kde')
```

```python
#hex: 산점도와 정보들을 육각형으로 표시
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind = 'hex')
```

```python
#hue: 기준점 세우고 색깔 구분
sns.jointplot(x='petal_length', y='petal_width', hue = 'species', data=iris)
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', hue = 'species', data=iris)
```

```python
#상관관계에 대해 heatmap 그리기
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

```python
#annot=True 하면 수치가 표시됨
sns.heatmap(iris.corr(), annot=True)
```

```python
#색 바꾸기 cmap
sns.heatmap(iris.corr(), annot=True, cmap='Blues')
```

```python
sns.heatmap(iris.corr(), annot=True, cmap='Reds')
```

```python
sns.heatmap(iris.corr(), annot=True, cmap='Greens')
```

```python
#산점도 scatterplot
sns.scatterplot(x = 'petal_length', y='petal_width', data=iris)
```

```python
#종 별로 점의 크기 지정 size
sns.scatterplot(x = 'petal_length', y='petal_width', size = 'species', data=iris)
```

```python
#색깔구분 relplot(hue='')
sns.relplot(x = 'petal_length', y='petal_width', hue='species', data=iris)
```

seaborn으로 데이터 시각화하기 2

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pairplot(tips)
```

```python
#hue : 조건 별로 색깔 구분
sns.pairplot(tips, hue='size')
```

```python
#height : 크기 변경
sns.pairplot(tips, hue='size', height=4)
```

```python
sns.pairplot(tips, hue='day')
```

```python
sns.pairplot(tips, hue='day', height=3)
```

```python
#막대 그래프 barplot
sns.barplot(x='day',y='tip',data=tips)
```

```python
#요일별로 카운트해서 값 분포 확인 value_counts()
tips.day.value_counts()
```

```python
#막대그래프 색 바꾸기 hue
sns.barplot(x='day',y='tip',hue='smoker',data=tips)
```

```python
#lmplot : 선형관계성
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
#hue 활용
sns.lmplot(x='total_bill', y='tip',hue='smoker', data=tips)
```

```python
#각각의 값을 구분지어 표시 col
sns.lmplot(x='total_bill', y='tip',hue='smoker',col='size', data=tips)
```

```python
#한 줄 당 그래프 개수 col_wrap
sns.lmplot(x='total_bill', y='tip',hue='smoker',col='size',col_wrap = 3, data=tips)
```

```python
#pivot 테이블 : 컬럼, 인덱스 지정해서 요약
pivot = tips.pivot_table(index='day', columns='smoker', values='tip')
pivot
```

```python
#heatmap 활용
sns.heatmap(pivot)
```

```python
#annot 옵션 활용
sns.heatmap(pivot, annot=True)
```

```python
#색 바꾸기 cmap
sns.heatmap(pivot, annot=True, cmap='Blues')
```

```python
#산점도 scatterplot
sns.scatterplot(x='total_bill',y='tip', data=tips)
```

```python
#size: 크기로 구분
sns.scatterplot(x='total_bill',y='tip', size='size',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip', size='smoker',data=tips)
```

```python
#relplot 활용
sns.relplot(x='total_bill',y='tip', data=tips)
```

```python
#색깔구분 hue
sns.relplot(x='total_bill',y='tip', hue='day', data=tips)
```

```python
#catplot : 카테코리 데이터로 산점도 그리기
sns.catplot(x='day', y='total_bill', data = tips)
```

```python
sns.catplot(x='smoker', y='total_bill', data = tips)
```
